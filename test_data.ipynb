Q4 - Locality

Classification model: 
    Modifications for Locality (PointNet++ inspired):
Point Sampling and Grouping:

First, we need to group points into neighborhoods. This can be achieved using methods like k-nearest neighbors (k-NN) or ball query.

In PointNet++, you apply a small PointNet-like architecture to each neighborhood.

Local Feature Extraction:

For each group of points (local neighborhood), you apply a local feature extraction process similar to the MLP layers you already have.

Aggregation:

After extracting features from each local neighborhood, you aggregate them using a symmetric function (like max pooling) to get a global representation of the point cloud

Explanation of Changes:
Neighborhood Grouping using K-NN:

We added a k-NN search using torch_cluster.knn, which finds the nearest k neighbors for each point.

This step allows us to consider local groups of points, which is the foundation of locality in PointNet++.

Local Feature Extraction:

After grouping the points into neighborhoods, we extract local features for each group. Each point now has access to its local neighborhood (k nearest neighbors).

We then use these local features to aggregate information, preserving local patterns in the data.

Max Pooling:

After extracting the local features, we apply max pooling to aggregate information across all points and neighborhoods, resulting in a fixed-size global feature representation.

Final Fully Connected Layers:

The global feature vector is then passed through fully connected layers to produce the final classification output.

Notes:
Neighborhood Size (k): You can experiment with the value of k, which defines how many nearest neighbors will be considered for each point.

Max Pooling: The use of max pooling helps aggregate local features into a global representation. This is similar to the original PointNet architecture, but now with locality taken into account.

Key Points:
PointNet++ is a hierarchical model that processes the point cloud in localized regions (neighborhoods).

This model still preserves the simplicity of PointNet while adding locality for better performance, especially for tasks like segmentation where local features are important.

You can further experiment with ball query (instead of k-NN) or multi-scale feature aggregation to enhance locality processing.
